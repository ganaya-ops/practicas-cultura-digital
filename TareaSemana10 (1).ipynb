### ***Ejercicio 1:***  
Pide dos números al usuario y muestre en pantalla el resultado de multiplicarlos.

````python
#Pedimos al usuario el primer número usando input() y lo convertimos a entero usando int(). Lo guardamos en la variable num1.
num1 = int(input("Dame el primer número: "))

#Pedimos al usuario el segundo número usando input() y lo convertimos a entero usando int(). Lo guardamos en la variable num2.
num2 = int(input("Dame el segundo número: "))

#Calculamos el producto de los dos números y lo guardamos en la variable producto.
producto = num1 * num2

#Mostramos en pantalla el resultado de la multiplicación usando print().
print(f"El resultado de la multiplicación es: {producto}")
````

### ***Ejercicio 2***:

Pida el nombre del usuario y su edad; luego muestre el mensaje: "Hola **[Nombre]** tienes **[años]** años!"

Cambia **[Nombre]** por el nombre que te dio el usuario y **[años]** por los años que te dio el usuario.

````python
#Pedimos al usuario su nombre usando input(). Lo guardamos en la variable nombre.
nombre = input("¿Cuál es tu nombre? ")

#Pedimos al usuario su edad usando input() y lo convertimos a entero usando int(). Lo guardamos en la variable edad.
edad = int(input("¿Cuántos años tienes? "))

#Mostramos el mensaje formateado con el nombre y la edad usando print() y f-strings.
print(f"Hola {nombre}, tienes {edad} años!")
````

### ***Ejercicio 3:***
Pide al usuario su nombre y apellido, y muestra el nombre completo.

````python
#Pedimos al usuario su nombre usando input(). Lo guardamos en la variable nombre.
nombre = input("Por favor, introduce tu nombre: ")

#Pedimos al usuario su apellido usando input(). Lo guardamos en la variable apellido.
apellido = input("Por favor, introduce tu apellido: ")

#Concatenamos el nombre y el apellido para formar el nombre completo. Lo guardamos en la variable nombre_completo.
nombre_completo = nombre + " " + apellido

#Mostramos el nombre completo en pantalla usando print().
print(f"Tu nombre completo es: {nombre_completo}")
````

### ***Ejercicio 4:***
Pide al usuario el precio de un producto y la cantidad que va a comprar, luego muestra el total a pagar.

````python
#Pedimos al usuario el precio del producto usando input() y lo convertimos a flotante usando float(). Lo guardamos en la variable precio.
precio = float(input("Introduce el precio del producto: "))

#Pedimos al usuario la cantidad que va a comprar usando input() y lo convertimos a entero usando int(). Lo guardamos en la variable cantidad.
cantidad = int(input("Introduce la cantidad que vas a comprar: "))

#Calculamos el total a pagar multiplicando el precio por la cantidad. Lo guardamos en la variable total_a_pagar.
total_a_pagar = precio * cantidad

#Mostramos el total a pagar en pantalla usando print() y f-strings, formateando a dos decimales.
print(f"El total a pagar es: ${total_a_pagar:.2f}")
````

### ***Ejercicio 5:***
Pide al usuario una cantidad de minutos y muestra cuántos segundos equivalen.

````python
#Pedimos al usuario la cantidad de minutos usando input() y la convertimos a entero usando int(). Lo guardamos en la variable minutos.
minutos = int(input("Introduce una cantidad de minutos: "))

#Convertimos los minutos a segundos multiplicándolos por 60 (ya que 1 minuto = 60 segundos). Lo guardamos en la variable segundos.
segundos = minutos * 60

#Mostramos la equivalencia en segundos en pantalla usando print() y f-strings.
print(f"{minutos} minutos equivalen a {segundos} segundos.")
````
